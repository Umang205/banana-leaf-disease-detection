```python
# ============================================================
# BANANA LEAF DISEASE DETECTION USING CNN
# COMPLETE GOOGLE COLAB PROJECT CODE
# ============================================================

# ============================================================
# STEP 1 - IMPORT LIBRARIES
# ============================================================

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
import zipfile

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D,
    MaxPooling2D,
    Flatten,
    Dense,
    Dropout,
    Input
)

from tensorflow.keras.preprocessing import image

# ============================================================
# STEP 2 - EXTRACT ZIP DATASET
# ============================================================

zip_path = "/content/archive1.zip"
extract_path = "/content/dataset"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Dataset Extracted Successfully!")

# ============================================================
# STEP 3 - DATASET PATH
# ============================================================

dataset_path = "/content/dataset/BananaLSD/AugmentedSet"

# ============================================================
# STEP 4 - CHECK DATASET CLASSES
# ============================================================

print("Dataset Classes:")
print(os.listdir(dataset_path))

# ============================================================
# STEP 5 - IMAGE SIZE
# ============================================================

img_width = 128
img_height = 128

# ============================================================
# STEP 6 - DATA PREPROCESSING
# ============================================================

datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    zoom_range=0.3,
    horizontal_flip=True,
    shear_range=0.3,
    brightness_range=[0.8, 1.2],
    validation_split=0.2
)

# ============================================================
# STEP 7 - LOAD TRAINING DATA
# ============================================================

training_set = datagen.flow_from_directory(
    dataset_path,
    target_size=(img_width, img_height),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

# ============================================================
# STEP 8 - LOAD VALIDATION DATA
# ============================================================

validation_set = datagen.flow_from_directory(
    dataset_path,
    target_size=(img_width, img_height),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

# ============================================================
# STEP 9 - PRINT CLASS INDICES
# ============================================================

print("Class Indices:")
print(training_set.class_indices)

# ============================================================
# STEP 10 - CLASS LABELS
# ============================================================

classes = [
    "cordana",
    "healthy",
    "pestalotiopsis",
    "sigatoka"
]

# ============================================================
# STEP 11 - BUILD CNN MODEL
# ============================================================

model = Sequential([

    # INPUT LAYER
    Input(shape=(128,128,3)),

    # FIRST CONVOLUTION LAYER
    Conv2D(
        32,
        (3,3),
        activation='relu'
    ),

    # FIRST POOLING LAYER
    MaxPooling2D(pool_size=(2,2)),

    # SECOND CONVOLUTION LAYER
    Conv2D(
        64,
        (3,3),
        activation='relu'
    ),

    # SECOND POOLING LAYER
    MaxPooling2D(pool_size=(2,2)),

    # THIRD CONVOLUTION LAYER
    Conv2D(
        128,
        (3,3),
        activation='relu'
    ),

    # THIRD POOLING LAYER
    MaxPooling2D(pool_size=(2,2)),

    # FLATTEN LAYER
    Flatten(),

    # DENSE LAYER
    Dense(
        128,
        activation='relu'
    ),

    # DROPOUT LAYER
    Dropout(0.5),

    # OUTPUT LAYER
    Dense(
        4,
        activation='softmax'
    )
])

# ============================================================
# STEP 12 - COMPILE MODEL
# ============================================================

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# ============================================================
# STEP 13 - MODEL SUMMARY
# ============================================================

model.summary()

# ============================================================
# STEP 14 - TRAIN MODEL
# ============================================================

history = model.fit(
    training_set,
    epochs=15,
    validation_data=validation_set
)

# ============================================================
# STEP 15 - SAVE MODEL
# ============================================================

model.save("banana_leaf_disease_model.keras")

print("Model Saved Successfully!")

# ============================================================
# STEP 16 - PLOT ACCURACY GRAPH
# ============================================================

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')

plt.legend(['Train', 'Validation'])

plt.show()

# ============================================================
# STEP 17 - PLOT LOSS GRAPH
# ============================================================

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')

plt.legend(['Train', 'Validation'])

plt.show()

# ============================================================
# STEP 18 - UPLOAD TEST IMAGE
# ============================================================

from google.colab import files

uploaded = files.upload()

# ============================================================
# STEP 19 - LOAD TEST IMAGE
# ============================================================

img_path = list(uploaded.keys())[0]

img = image.load_img(
    img_path,
    target_size=(128,128)
)

# ============================================================
# STEP 20 - DISPLAY TEST IMAGE
# ============================================================

plt.imshow(img)
plt.axis("off")
plt.show()

# ============================================================
# STEP 21 - IMAGE PREPROCESSING
# ============================================================

img_array = image.img_to_array(img)

img_array = np.expand_dims(img_array, axis=0)

img_array = img_array / 255.0

# ============================================================
# STEP 22 - PREDICT IMAGE
# ============================================================

prediction = model.predict(img_array)

print("Prediction Probabilities:")
print(prediction)

# ============================================================
# STEP 23 - GET FINAL PREDICTION
# ============================================================

predicted_class = classes[np.argmax(prediction)]

print("Predicted Disease:", predicted_class)

# ============================================================
# STEP 24 - DOWNLOAD TRAINED MODEL
# ============================================================

files.download("banana_leaf_disease_model.keras")
```
